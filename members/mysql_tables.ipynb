```python
import mysql.connector

# إعداد الاتصال بقاعدة البيانات MySQL
db_config = {
    'host': 'localhost',
    'database': 'sportx_saas_trail_deletet',
    'user': 'laravel_user',
    'password': 'Askar@1984'
}

# الاتصال بقاعدة البيانات
conn = mysql.connector.connect(**db_config)
cursor = conn.cursor()

# دالة لاستخراج الأعمدة من جدول معين
def get_columns(table_name):
    cursor.execute(f"DESCRIBE {table_name}")
    columns = cursor.fetchall()
    column_names = [column[0] for column in columns]  # استخراج اسم العمود
    return column_names

# استخراج الأعمدة للجدولين
client_details_columns = get_columns('quotations')
client_contacts_columns = get_columns('client')

# حفظ الأعمدة في ملف نصي
with open('columns_info.txt', 'w') as file:
    file.write("الأعمدة لجدول client_details:\n")
    file.write(", ".join(client_details_columns) + "\n\n")
    
    file.write("الأعمدة لجدول client_contacts:\n")
    file.write(", ".join(client_contacts_columns) + "\n")

# غلق الاتصال بقاعدة البيانات
cursor.close()
conn.close()

print("تم حفظ الأعمدة في ملف columns_info.txt")
```
